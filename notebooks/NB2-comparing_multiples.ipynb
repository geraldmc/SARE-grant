### Comparing Multiple Plots

```python
%run ../lib/spectral_lib.py
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
# Data path relative to this location.
base_path = '../data/imagesets'

image_list = []
sample_list = []
grayscale_list = []

NDVI_list = []
GNDVI_list = []
NDRE_list = []

ymd1 = '/2017-04-25/'
ymd2 = '/2017-05-10/'
ymd3 = '/2017-05-25/'
ymd4 = '/2017-07-28/'
ymd5 = '/2017-09-10/'
ymd6 = '/2017-11-13/'
```

```python
# Early Treatments 1, 2, 3
image_list.append((ymd1, '002_1'))
image_list.append((ymd2, '002_1')) 
image_list.append((ymd3, '002_1')) 
image_list.append((ymd1, '004_2')) 
image_list.append((ymd2, '004_2')) 
image_list.append((ymd3, '004_2')) 
image_list.append((ymd1, '001_3')) 
image_list.append((ymd2, '001_3')) 
image_list.append((ymd3, '001_3'))
```

```python
for idx in range(len(image_list)):
    sample_list.append(get_path2(base_path, image_list[idx][0], image_list[idx][1]))
for idx in range(len(sample_list)):
    grayscale_list.append(get_greyscale(sample_list[idx][0], sample_list[idx][1], sample_list[idx][2]))
for idx in range(len(grayscale_list)):
    NDVI_list.append(calc_ndvi(grayscale_list[idx]))
    GNDVI_list.append(calc_gndvi(grayscale_list[idx]))
```

```python
plt.gcf().clear()

fig = plt.figure(figsize=(18, 5))

for idx in range(len(sample_list)):
    plt.subplot(int('33' + str(idx+1)))
    plt.title('NDVI ' + str(sample_list[idx][1]) + ' ' + sample_list[idx][2])
    plt.imshow(NDVI_list[idx], cmap=fastie_lut_cmap)
fig = plt.gcf()
#fig.savefig('../readme_resources/img/multiples_123_NDVI.png', format='png')
```

```python
plt.gcf().clear()

fig = plt.figure(figsize=(18, 5))

for idx in range(len(sample_list)):
    plt.subplot(int('33' + str(idx+1)))
    plt.title('GNDVI ' + str(sample_list[idx][1]) + ' ' + sample_list[idx][2])
    plt.imshow(GNDVI_list[idx], cmap=fastie_lut_cmap)
fig = plt.gcf()
#fig.savefig('../readme_resources/img/multiples_123_NDVI.png', format='png')
```
